```python
dataset_root_dir = '/home/gaurav/PycharmProjects/code-mixed/data'
dataset_type = ['train']
combination_type = [0,1]
files = []
for d_type in dataset_type:
    for c_type in combination_type:
        path = pathlib.Path(f'{dataset_root_dir}/{d_type}_hinglish_uncased/{c_type}')
        p = path.glob('**/*')
        files.extend([x for x in p if x.is_file()])
files = list(set(files))
print(len(files))
```

```python
files[0]
```

```python
files = [str(file) for file in files]
```

```python
flist = ','.join(files)
```

```python
custom_symbols = [text.transform.FLD, 
                text.transform.TK_MAJ,
                text.transform.TK_UP,
                text.transform.TK_REP,
                text.transform.TK_WREP]
```

```python
str_specialcases = ",".join(custom_symbols)
```

```python
str_specialcases
```

```python
spm.SentencePieceTrainer.Train(f'--input={flist} --model_prefix=hinglish_spm --vocab_size=60000 --input_sentence_size=22500000 --unk_id=0 --bos_id=1 --eos_id=2 --pad_id=3 --unk_piece={text.transform.UNK} --bos_piece={text.transform.BOS} --eos_piece={text.transform.EOS} --pad_piece={text.transform.PAD} --user_defined_symbols={str_specialcases}')
```

```python
sp = spm.SentencePieceProcessor()
```

```python
sp.Load("hinglish_spm.model")
```
